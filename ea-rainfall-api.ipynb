```python
# Import libraries required
import requests
import pandas as pd

r = requests.get('https://environment.data.gov.uk/flood-monitoring/id/stations?parameter=rainfall')
j = r.json()

# Read through the response to the above url, which is a list of all rainfall sites and their "measures"
# create a data frame with a line for each "measure" (i.e. time series) and add the site metadata to this line
# result is a data frame for every rainfall time series

ts=[]
stationKeys = ['stationReference', 'notation', 'easting','northing','long','lat','@id','dateOpened', 'wiskiID' ]

for itm in j['items']:
    for m in itm['measures']:
        ts.append(m)
        subdict = {k: itm[k] for k in stationKeys if k in itm}
        ts[(len(ts)-1)].update(subdict)

df=pd.DataFrame(ts)
# Print the length (number of stations)
print(len(df))
# print the first 5 lines
df.head()
```
